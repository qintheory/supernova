```python
# Definite molecule types - the keys and values
node_c = {"id": 0, "atom": "C", "size": 12, "color": "#3d3d3d"}
node_o = {"id": 0, "atom": "O", "size": 16, "color": "#3d3d3d"}
node_n = {"id": 0, "atom": "N", "size": 14, "color": "#7093cc"}
node_h = {"id": 0, "atom": "H", "size": 1, "color": "#3d3d3d"}
```

```python
# Creating one unite of nodes

nodes = [ ]

for i in range(c+o+h+n):
    if i < c: 
        node_c["id"] = i  
        nodes.append(node_c)
    else: 
        if i < (o+c):
            node_o["id"] = i
            nodes.append(node_o)
        else: 
            if i < (o+c+n):
                node_n["id"] = i
                nodes.append(node_n)
            else: 
                if i < (o+c+n+h):
                    node_h["id"] = i
                    nodes.append(node_h)

print(nodes)
```
